### Now we can see that Price column still consists of more %ge(21.71%) of Null values but Price is very important column for Interpretaion hence we cannot drop the whole Price column, but here we can remove the rows which has Price values = Null or NaN

### To do this we Have 2 methods in Python

### Method 1: -

```python
# lets first describe the Price column here we can see that we cannot 
# Impute the Null values of Price with other values of the column, hence we remove:


df['Price'].describe()
```

```python
df['Price'].notnull()
```

```python
# here lets pass this as Index value to our dataset 'df', but let's not update df yet:

df[df['Price'].notnull()]
```

### Method 2: -

```python
[~np.isnan(df['Price'])]
```

### Now using any method 1 or 2 we can update our dataset 'df' finally

```python
df = df[df['Price'].notnull()]
df
```

```python
# here we can see the dataset having no Null values in Price column

round(df.isnull().sum().sort_values(ascending = False)/len(df) * 100, 2)
```

### Here Landsize also has little high Null Values but since it is also an important column for us so just like Price column we can drop the rows containing Null values in Landsize

```python
# lets first describe the Landsize column where we can see that we cannot 
# Impute the Null values of Landsize with other values of the column, hence we remove:

df['Landsize'].describe()
```

```python
df = df[df['Landsize'].notnull()]
df
```

### As we can see Lattitude and Longitude has very few Null values, so rather dropping the rows we can Impute the Null values with Mean of the values of individual attributes.

```python
df.loc[:, ['Lattitude', 'Longtitude']].describe()  # here we can describe the mean values first before imputing
```

```python
df['Lattitude'].mean()
```

```python
df.loc[:, 'Lattitude'].fillna(df['Lattitude'].mean(), inplace = True)  
# we can use inplace fn to update 'df' instead updating and the assignining it using 
# df = df.loc[:, 'Lattitude'].fillna(df['Lattitude'].mean()), it means same
```

```python
df['Longtitude'].mean()
```

```python
df.loc[:, 'Longtitude'].fillna(df['Longtitude'].mean(), inplace = True)
```

```python
round(df.isnull().sum().sort_values(ascending = False)/len(df) * 100, 2)
```

### Now let us describe Car and Bathroom columns and clean their data also

```python
df.loc[:, ['Car', 'Bathroom']].describe()  # here we can describe the mean values first before imputing
```

### here above we can clearly see that mean of both Car and Bathroom columns are between 1 and 2, so we can replace/impute the Null values of Car and Bathroom either by 1 or 2to decide on either to impute with 1 or 2, we can see our data once which has highest values in the table is it 1 or 2we can use Count for this

### Method 1: -

```python
df.groupby('Car').Car.count().sort_values(ascending = False)  # here it is clear that 2 is the highest assigned value
```

### Method 2: -

```python
df['Car'].astype('category').value_counts()
```

```python
# |||ly we can Count for Bathroom column also

df.groupby('Bathroom').Bathroom.count().sort_values(ascending = False) # here it is clear that 1 is the highest assigned value
```

```python
df.loc[:, 'Car'].fillna(2, inplace = True)
```

```python
df.loc[:, 'Bathroom'].fillna(1, inplace = True)
```

```python
round(df.isnull().sum().sort_values(ascending = False)/len(df) * 100, 2)
```
